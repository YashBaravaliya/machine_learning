### Q.1) Explain the following with example
1. Artificial intelligence
2. Machine Learning
3. Deep Learning

1. Artificial Intelligence (AI): Artificial Intelligence refers to the development of intelligent machines that can perform tasks typically requiring human intelligence. AI systems can perceive their environment, reason about it, and make decisions or take actions accordingly. One example of AI is virtual assistants like Amazon's Alexa or Apple's Siri. These assistants use natural language processing and machine learning algorithms to understand and respond to user queries, providing relevant information or performing tasks like setting reminders or playing music.

2. Machine Learning (ML): Machine Learning involves the development of algorithms and models that enable computers to learn from data and make predictions or decisions without explicit programming. An example of machine learning is email spam filters. By analyzing large sets of labeled emails (spam and non-spam), ML algorithms can learn patterns and features that distinguish between the two categories. Once trained, the spam filter can accurately classify incoming emails as either spam or legitimate based on its learned knowledge.

3. Deep Learning (DL): Deep Learning is a subset of machine learning that focuses on training artificial neural networks with multiple layers, known as deep neural networks. DL algorithms excel at processing complex data like images, sounds, or natural language. For instance, image recognition is a popular application of deep learning. Deep neural networks can be trained on vast collections of labeled images to recognize and classify objects in new images. An example is the use of deep learning in self-driving cars, where the system analyzes real-time camera inputs to identify objects like pedestrians, traffic signs, or other vehicles.

### Q.2) What is supervised learning? List some example of supervised learning.

Supervised learning is the types of machine learning in which machines are trained using well "labelled" training data, and on basis of that data, machines predict the output. The labelled data means some input data is already tagged with the correct output.



Email Spam Classification: Given a dataset of labeled emails (spam or non-spam), a supervised learning algorithm can be trained to classify new emails as either spam or non-spam based on patterns and features extracted from the email content.

Image Classification: Supervised learning can be used to classify images into different categories. For example, a model can be trained on a dataset of labeled images of animals (cats, dogs, birds, etc.) and then be able to classify new images of animals based on what it has learned.

Sentiment Analysis: In sentiment analysis, supervised learning can be used to determine the sentiment or emotion expressed in a piece of text. By training on a dataset of labeled texts (positive, negative, or neutral sentiment), a model can be built to predict the sentiment of new, unseen texts.

Fraud Detection: Supervised learning algorithms can be trained to identify fraudulent transactions by learning from historical data that includes labeled examples of fraudulent and non-fraudulent transactions. The model can then be used to detect potential fraud in real-time transactions.

Medical Diagnosis: Supervised learning can be applied to medical data to assist in the diagnosis of diseases. By training on labeled medical records and corresponding diagnoses, a model can learn patterns and symptoms indicative of specific diseases, aiding doctors in making accurate diagnoses.

Predictive Maintenance: Supervised learning can be used to predict equipment failures or maintenance needs. By training on historical data that includes labeled instances of equipment failure or maintenance requirements, a model can be built to predict when similar failures or maintenance needs might occur in the future.

### Q.3) What is unsupervised learning? List some example of unsupervised learning.

As the name suggests, unsupervised learning is a machine learning technique in which models are not supervised using training dataset. Instead, models itself find the hidden patterns and insights from the given data. It can be compared to learning which takes place in the human brain while learning new things. It can be defined as:

Unsupervised learning is a type of machine learning in which models are trained using unlabeled dataset and are allowed to act on that data without any supervision



Clustering: Unsupervised learning algorithms can be used to group similar data points together based on their inherent similarities or patterns. For example, clustering can be applied to customer segmentation in marketing, grouping customers with similar purchasing behaviors or demographics.

Anomaly Detection: Unsupervised learning can be utilized to detect unusual or anomalous patterns in data. This can be useful in identifying fraudulent activities, network intrusions, or manufacturing defects that deviate from the norm.

Dimensionality Reduction: Unsupervised learning techniques like Principal Component Analysis (PCA) or t-SNE (t-Distributed Stochastic Neighbor Embedding) can be employed to reduce the dimensionality of a dataset while retaining meaningful information. This can help visualize and understand high-dimensional data or improve computational efficiency in subsequent tasks.

Market Basket Analysis: Unsupervised learning can be applied to analyze transactional data, such as in a retail setting, to identify associations and patterns among items frequently purchased together. This information can be utilized for product placement, recommendations, and pricing strategies.

Topic Modeling: Unsupervised learning techniques, such as Latent Dirichlet Allocation (LDA), can be used to identify underlying topics or themes within a collection of documents. This is particularly useful in tasks like text mining, content analysis, and organizing large document sets.

Generative Models: Unsupervised learning can involve training generative models such as Generative Adversarial Networks (GANs) or Variational Autoencoders (VAEs). These models can learn to generate new, realistic samples that resemble the training data. Applications range from generating synthetic images to creating music or generating text.

### Q.5) what are the main different between supervised unsupervised and semi-supervised learning?

Supervised Learning:

Training Data: Supervised learning requires labeled training data, where each data instance is paired with its corresponding target or output label.
Learning Goal: The goal of supervised learning is to learn a mapping between input features and their associated output labels. It aims to make predictions or classify new, unseen instances based on the learned patterns from labeled data.
Example: Email spam classification, where the algorithm learns to classify emails as spam or non-spam based on labeled training examples.

Unsupervised Learning:

Training Data: Unsupervised learning uses unlabeled training data, where only input features are available without corresponding output labels.
Learning Goal: The goal of unsupervised learning is to discover patterns, structures, or relationships within the data. It involves clustering similar instances together or finding hidden representations in the data.
Example: Customer segmentation, where the algorithm groups customers into segments based on their similarities in purchasing behavior or demographics.

Semi-Supervised Learning:

Training Data: Semi-supervised learning involves a combination of labeled and unlabeled training data, where a portion of the data has associated output labels, and the rest is unlabeled.
Learning Goal: The goal of semi-supervised learning is to leverage both labeled and unlabeled data to improve the learning process. It aims to utilize the labeled data to guide and enhance the discovery of patterns or structures in the unlabeled data.
Example: Speech recognition, where a small amount of transcribed speech data (labeled) is used alongside a large amount of unlabeled speech data to improve the accuracy of the speech recognition system.


### Q.6) What is train test and validation split? Explain the importance of each term.

- Training set: It is essential for the model to learn and adjust its parameters based on labeled examples, enabling it to capture the patterns and relationships in the data.

- Test set: It is crucial for evaluating the model's performance on unseen data and estimating its ability to generalize to real-world scenarios. It helps identify issues like overfitting, where the model performs well on the training data but poorly on new data.

- Validation set: It aids in fine-tuning the model and optimizing its performance by allowing comparisons between different models or parameter settings. It provides an independent measure of performance during the development process and helps in making decisions to improve the model's effectiveness.

### Q.7) How can unsupervised learning be used in anomaly detection?

Unsupervised learning is commonly used in anomaly detection because it allows the model to learn patterns and structures in the data without relying on labeled examples of anomalies. Here's how unsupervised learning can be applied in anomaly detection:

1. Data Representation: First, the data needs to be appropriately represented for unsupervised learning. This often involves feature engineering to extract relevant information or transform the data into a suitable format for analysis.

2. Unsupervised Learning Algorithm: Various unsupervised learning algorithms can be employed for anomaly detection, such as clustering algorithms, density estimation methods, or dimensionality reduction techniques.

   - Clustering: Clustering algorithms can group similar instances together based on their proximity in feature space. Anomalies are identified as instances that do not belong to any cluster or belong to a small or sparse cluster.

   - Density Estimation: Density-based methods estimate the underlying data distribution and identify anomalies as instances that have a low probability or fall outside regions of high-density data.

   - Dimensionality Reduction: Dimensionality reduction techniques, like PCA or autoencoders, can be used to transform high-dimensional data into a lower-dimensional representation. Anomalies can then be detected as instances that deviate significantly from the reconstructed or projected data.

3. Anomaly Score Calculation: Once the unsupervised learning model is trained, it assigns anomaly scores to each instance in the data. The anomaly score represents the likelihood or degree of abnormality for each instance based on the learned patterns or structure.

4. Threshold Setting: A threshold is chosen to distinguish between normal and anomalous instances based on the anomaly scores. Instances with scores above the threshold are classified as anomalies.

5. Evaluation and Refinement: The performance of the anomaly detection system is evaluated using labeled data (if available) or through domain experts' feedback. The threshold or model parameters can be adjusted iteratively to achieve a balance between detection accuracy and false positive/negative rates.
